# Read all images in target folder

```python
original_x = 30720
original_y = 23040

true_x = 2048
true_y = 1536

root_dir = r"C:\Users\alisa\Documents\Jupyter\Coral_dataset\JohnReed\Photos - CPCe and Density"

# Lists to store filenames by type
ara_files = []
cpc_files = []
jpg_files = []

# Traverse all subdirectories and files
for subdir, _, files in os.walk(root_dir):
    # print(subdir)
    if "Removed" in subdir:
        continue
    for file in files:
        # Get the full path and file extension
        full_path = os.path.join(subdir, file)
        file_extension = file.lower().split('.')[-1]  # Extract file extension (case-insensitive)
        
        # Append files to corresponding lists based on their extension
        if file_extension == "ara":
            ara_files.append(full_path)
        elif file_extension == "cpc":
            cpc_files.append(full_path)
        elif file_extension == "jpg":
            jpg_files.append(full_path)
```

```python
# Sort each list alphabetically
ara_files.sort()
cpc_files.sort()
jpg_files.sort()

# Print results
print("\nCPC Files:")
print(cpc_files)
print(len(cpc_files))
print(len(jpg_files))
```

```python
# Filter files with the same base name
def filter_matching_files(cpc_files, jpg_files):
    # Extract base filenames (without extensions)
    cpc_basenames = {os.path.splitext(os.path.basename(file))[0] for file in cpc_files}
    jpg_basenames = {os.path.splitext(os.path.basename(file))[0] for file in jpg_files}

    # Find common base filenames
    matching_basenames = cpc_basenames.intersection(jpg_basenames)

    # Filter the original lists
    filtered_cpc_files = [file for file in cpc_files if os.path.splitext(os.path.basename(file))[0] in matching_basenames]
    filtered_jpg_files = [file for file in jpg_files if os.path.splitext(os.path.basename(file))[0] in matching_basenames]

    return filtered_cpc_files, filtered_jpg_files

# Example usage
filtered_cpc_files, filtered_jpg_files = filter_matching_files(cpc_files, jpg_files)

# Print results
print("Filtered CPC Files:")
print(filtered_cpc_files)
print(len(filtered_cpc_files))
print("\nFiltered JPG Files:")
print(filtered_jpg_files)
print(len(filtered_jpg_files))
```

```python
details_path = r".\dataset\CPCe Code Names- used for Pulley Ridge 2012-2015 FINAL .TXT"

f = open(details_path)
lines = f.readlines()
num_class = int(lines[2])
print(num_class)
classes = lines[3:3 + num_class]
label_names = lines[3 + num_class + 1:]
f.close()

# Initialize ds1 and ds2 as pandas DataFrames
ds1 = pd.DataFrame(columns=["class_name", "class_number", "explanation"])  # For classes
ds2 = pd.DataFrame(columns=["name", "explain_name", "class_name"])  # For labels

# Populate ds1
idx = 0
for cl in classes:
    label1 = cl.split(",")
    class_name = label1[0].replace('\"', '')
    explanation = label1[1].replace('\"', '').strip()
    ds1 = pd.concat([ds1, pd.DataFrame([[class_name, idx, explanation]], columns=ds1.columns)], ignore_index=False)
    idx += 1

# Populate ds2
for lbl in label_names:
    label1 = lbl.split(",")
    name = label1[0].replace('\"', '')
    explain_name = label1[1].replace('\"', '')
    class_name = label1[2].replace('\"', '').strip()
    ds2 = pd.concat([ds2, pd.DataFrame([[name, explain_name, class_name]], columns=ds2.columns)], ignore_index=False)

# Perform the join
full_ds = pd.merge(ds2, ds1, on="class_name", how="inner")

full_ds.index = full_ds.index + 1

full_ds.to_csv(r".\dataset\CPCe_Code.csv")
```

# Convert CPCe format to label csv format

```python
def read_label(cpc_file, class_dataset):
    # Open the file and read lines
    with open(cpc_file, 'r') as f:
        lines = f.readlines()
    
    # Extract dimensions and number of points
    line1 = lines[0]
    data = lines[6:56]
    label = lines[56:106]
    new_v = line1.split(",") 
    demision_length = int(new_v[2])
    demision_width = int(new_v[3])
    count_points = int(lines[5])

    # Prepare the output dataset
    dataset_out = []

    # Process coordinate data
    for n in range(50):
        data1 = data[n].split(",")
        x = int(int(data1[0]) * 2048 / 30720)  # Rescale x
        y = int(int(data1[1]) * 1536 / 23040)  # Rescale y
        
        # Process label data
        label1 = label[n].split(",")
        label_name = label1[1].replace('\"', '').strip()
        class_name = class_dataset.loc[class_dataset['name'] == label_name, 'class_name'].values[0]
        
        # Append to dataset
        dataset_out.append({
            "Number": n + 1,
            "x": x,
            "y": y,
            "label_name": label_name,
            "class_name": class_name
        })

    # Convert to DataFrame
    dataset_out = pd.DataFrame(dataset_out)

    return dataset_out
```

```python
destination_dir = "./dataset/labels_csv"
for cpc_file in filtered_cpc_files:
    dataset_out = read_label(cpc_file, full_ds)
    name_txt = os.path.basename(cpc_file)
    destination_path = os.path.join(destination_dir, name_txt[:-4] + ".csv")
    dataset_out.to_csv(destination_path, index=False)
```
